- Name : 게임의 이름입니다.

- Platform : 게임이 지원되는 플랫폼의 이름입니다.

- Year : 게임이 출시된 연도입니다.

- Genre : 게임의 장르입니다.

- Publisher : 게임을 배급한 회사입니다.

- NA_Sales : 북미지역에서의 출고량입니다.

- EU_Sales : 유럽지역에서의 출고량입니다.

- JP_Sales : 일본지역에서의 출고량입니다.

- Other_Sales : 기타지역에서의 출고량입니다.

```python
#23보다 크거나 100보다 작은 수들은 1900년대로 간주, 년도 0000 형식으로 바꿔줌
large_year = games_real[(games_real['Year'] > 23) & (games_real['Year'] <100)]
large_year
```

```python
large_year['Year'] = large_year['Year'].astype(int)
large_year
```

```python
large_year['Year'] = large_year['Year'] + 1900
large_year
```

```python
largeyear = games_real[(games_real['Year'] > 23) & (games_real['Year'] < 100)].index
games_real.drop(largeyear, inplace = True)
```

```python
#Name, Year, Genre, Publisher, Sales들이 모두 같은 값은 같은 데이터셋으로 간주
games_final_set = games_final[games_final.duplicated(subset=['Name', 'Year', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], keep=False)].sort_values(by=['Name'])
games_final_set
```

```python
#중복되는 데이터는 첫번째 값만 남기고 삭제
games_final_set.drop_duplicates(subset=['Name', 'Year', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], keep='first', inplace=True)
```

```python
games_final_set2 = games_final[games_final.duplicated(subset=['Name', 'Year', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], keep=False)].sort_values(by=['Name']).index
games_final.drop(games_final_set2, inplace=True)
```

```python
games_final_set
```

```python
games_final2 = pd.merge(games_final, games_final_set, how='outer')
games_final2.reset_index()
```

```python
#Name, Year, Publisher이 같은 데이터는 같은 게임의 데이터가 나눠져있다고 간주
x = games_final2[games_final2.duplicated(subset=['Name', 'Year', 'Publisher'], keep=False)].sort_values(by=['Name'])
```

```python
x.head(20)
```

```python
x2 = games_final2[games_final2.duplicated(subset=['Name', 'Year','Publisher'], keep=False)].index
games_final2.drop(x2, inplace = True)
```

```python
games_final2.sort_values(by=['Name']).head(20)
```

```python
#Name, Year, Genre, Publisher가 같은 데이터는 하나의 게임이되 여러 데이터로 나누어져 있다고 간주
#각각의 Sales를 합쳐줌
x3 = x.groupby(['Name', 'Year', 'Genre', 'Publisher']).sum()
```

*1. 지역별로 선호하는 게임 장르*

```python
#각 지역별 Genre에 따른 Sales 분석
Genre_country = games_final3.groupby('Genre')['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'].mean()
Genre_country.reset_index()
```

```python
#NA 지역에서 가장 많이 팔린 상위 3개의 게임 장르
genre_na = Genre_country.groupby('Genre')['NA_Sales'].mean().nlargest(3)
```

```python
genre_na.reset_index()
```

```python
#EU 지역에서 가장 많이 팔린 상위 3개의 게임 장르
genre_eu = Genre_country.groupby('Genre')['EU_Sales'].mean().nlargest(3)
genre_eu
```

```python
#JP 지역에서 가장 많이 팔린 상위 3개의 게임 장르
genre_jp = Genre_country.groupby('Genre')['JP_Sales'].mean().nlargest(3)
genre_jp
```

```python
#Other 지역에서 가장 많이 팔린 상위 3개의 게임 장르
genre_other = Genre_country.groupby('Genre')['Other_Sales'].mean().nlargest(3)
genre_other
```

```python
#지역별 선호하는 게임 장르 분석 시각화
import matplotlib.pyplot as plt

#NA 지역별 선호 장르 시각화
ratio = genre_na
labels = ['Platform', 'Shooter', 'Sports']
plt.figure(figsize=(12,12))
plt.subplot(221)
explodes =(0.1,0,0)
plt.pie(ratio, labels=labels, autopct='%.1f%%', colors=['bisque','darkorange','burlywood'], explode=explodes)
plt.title('North America')

#EU 지역별 선호 장르 시각화
ratio = genre_eu
labels = ['Shooter', 'Platform', 'Racing']
plt.subplot(222)
explodes =(0.1,0,0)
plt.pie(ratio, labels=labels, autopct='%.1f%%', colors=['darkorange', 'bisque', 'darkkhaki'], explode=explodes)
plt.title('Europe')

#JP 지역별 선호 장르 시각화
ratio = genre_jp
labels = ['Role-Playing', 'Platform', 'Fighting']
plt.subplot(223)
explodes =(0.1,0,0)
plt.pie(ratio, labels=labels, autopct='%.1f%%', colors=['olivedrab', 'bisque', 'darkseagreen'], explode=explodes)
plt.title('Japan')

#Other 지역 선호 장르 시각화
ratio = genre_other
labels = ['Shooter', 'Racing', 'Sports']
plt.subplot(224)
explodes =(0.1,0,0)
plt.pie(ratio, labels=labels, autopct='%.1f%%', colors=['darkorange', 'darkkhaki','burlywood'], explode=explodes)
plt.title('Other')

plt.suptitle('Prefer Genre by Country',fontsize=20)
plt.show()
```
