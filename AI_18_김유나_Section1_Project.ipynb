```python
#NA ÏßÄÏó≠ÏóêÏÑú Í∞ÄÏû• ÎßéÏù¥ ÌåîÎ¶∞ ÏÉÅÏúÑ 3Í∞úÏùò Í≤åÏûÑ Ïû•Î•¥
genre_na = Genre_country.groupby('Genre')['NA_Sales'].mean().nlargest(3)
```
